```python

# Apply a function to check the value of `class` column
# If class=1, marker_color value will be green
# If class=0, marker_color value will be red
# Create a function to determine marker color based on class value
def determine_marker_color(launch_site):
    # Get all launches for the site
    launches = spacex_df[spacex_df['Launch Site'] == launch_site]
    
    # Check the class values and determine the marker color
    if (launches['class'] == 1).any():
        return '#28a745'  # Green for at least one success
    else:
        return '#dc3545'  # Red for no successes (all failures)

# Apply the function to create a new column for marker_color
launch_sites_df['marker_color'] = launch_sites_df['Launch Site'].apply(determine_marker_color)
```

```python
# TASK 3: Calculate the distances between a launch site to its proximities
import pandas as pd
import numpy as np
import io
from js import fetch

# Define the Haversine function
def haversine(lat1, lon1, lat2, lon2):
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = np.sin(dlat / 2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2)**2
    c = 2 * np.arcsin(np.sqrt(a))
    r = 6371  # Radius of Earth in kilometers
    return c * r

# Fetch the CSV file from the URL
async def load_data():
    URL = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/spacex_launch_geo.csv'
    resp = await fetch(URL)
    spacex_csv_file = io.BytesIO((await resp.arrayBuffer()).to_py())
    return pd.read_csv(spacex_csv_file)

# Load the data
spacex_df = await load_data()

# Create launch_sites_df from spacex_df
launch_sites_df = spacex_df[['Launch Site', 'Lat', 'Long']].groupby(['Launch Site'], as_index=False).first()

# Calculate distances between launch sites
distances = []

for i, site1 in launch_sites_df.iterrows():
    for j, site2 in launch_sites_df.iterrows():
        if i != j:  # Avoid self-comparison
            distance = haversine(site1['Lat'], site1['Long'], site2['Lat'], site2['Long'])
            distances.append({
                'Launch Site 1': site1['Launch Site'],
                'Launch Site 2': site2['Launch Site'],
                'Distance (km)': distance
            })

# Create a DataFrame from the distances list
distances_df = pd.DataFrame(distances)

# Display the distances DataFrame
print(distances_df)
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
